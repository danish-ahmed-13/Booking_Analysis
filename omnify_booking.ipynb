### Datatype Conversion

```python
df['Duration (mins)'] = df['Duration (mins)'].astype(int)
df['Price'] = df['Price'].astype(int)
```

```python
df.info()
```

```python
df['revenue'] = df['Price'] * df['Duration (mins)']
```

```python
df.head(2)
```

### Extracting monnth from Booking Date for monthly analysis

```python
df['month'] = df['Booking Date'].dt.month
```

```python
df.head(2)
```

```python
monthly_revenue = df.groupby('month')['Price'].sum().sort_values(ascending=False)
monthly_revenue.plot(kind='bar', color='skyblue')
plt.title('Monthly Revenue Analysis')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.xticks(rotation=0)
plt.show()
```

### Total Revenue Calculation

```python
df['revenue'].sum()
print(f"Total Revenue: {df['revenue'].sum():,.2f}")
```

```python
df['revenue'].max()
print(f"Highest Revenue: {df['revenue'].max():,.2f}")
```

```python
df['Price'].max()
print(f"Highest price for single booking: {df['Price'].max():,.2f}")
```

```python
df.groupby('Status')['revenue'].sum().sort_values(ascending=False).plot(kind='bar', color='orange')
plt.title('Revenue by Booking Status')
plt.xlabel('Booking Status')
plt.ylabel('Total Revenue')
plt.show()
```

```python
df.groupby('month')['Booking ID'].count().plot(kind='bar', color='teal')
plt.title('Monthly Booking Count')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=0)
plt.show()
```

```python
df.groupby('Booking Type')['Booking ID'].count().plot(kind='bar', color='purple')
plt.title('Booking Count by Booking Type')
plt.xlabel('Booking Type')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.show()
```
